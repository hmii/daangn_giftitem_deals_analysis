```python


def category_rank(x):
    token = x.split()
    for i in token:
        token_list.append(i)
        
for a_cate in list(df.cate_final.unique()):     
    zipp ={}
    token_list = []
    temp = df[df.cate_final == a_cate]
    temp.token_sum.apply(category_rank)
    token_set = list(set(token_list))
    count_token = [token_list.count(i) for i in token_set]
    zipp['count']= dict(zip(token_set, count_token))
    pd.DataFrame(zipp).reset_index().sort_values(by='count', ascending = False).head(20)
```
