```python
import pandas as pd
import re
from datetime import date, datetime, timedelta
from dateutil.relativedelta import relativedelta
import os
import numpy as np
```

```python
def when(x):   
    p = re.compile('(\d+)')
    delta = -int(p.findall(x)[0])
    if "초" in x:
        result = timedelta(seconds = delta)
        
    elif "분" in x:
        result = timedelta(minutes = delta)
        
    elif "시간" in x:
        result = timedelta(hours = delta)
     
    elif "일" in x:
        result = timedelta(days = delta)
        
    elif "달" in x:
        result = relativedelta(months = delta)
        
    elif "년" in x:
        result = relativedelta(years = delta)

    return the_date + result
```

```python
# 폴더 내 리스트 뽑기 
folder_name = 'data_gu'
file_names = os.listdir(folder_name)
file_names
```

```python
# upload_date : 수집 당시 datetime을 기준으로 추적한 업로드 datetime
# the_date : 수집 당시 datetime
# the_date_ymd : 기준 날짜는 date만 추출 
# reup : 끌올여부 (끌올: 1 , no끌올: 0)
# region_gu : 구 추출 

total = pd.DataFrame()

for file_name in file_names:
    file_date = file_name[-22:-5]

    year = int(file_date[0:4])
    month = int(file_date[5:7])
    day = int(file_date[8:10])
    hour = int(file_date[11:13])
    minute = int(file_date[13:15])
    second = int(file_date[15:17])
    
    the_date = datetime(year, month, day, hour, minute, second)
    
    df = pd.read_excel(f'./{folder_name}/'+file_name)
    df = df[df.time.str.contains("초|분|시간")]
    df['upload_date'] = df['time'].apply(when)   
    # df['upload_ym_date'] = df['upload_date'].dt.strftime('%Y-%m') 
    df['the_date'] = the_date
    total = total.append(df)
    print(len(total))
    
total['the_date_ymd'] = total['the_date'].dt.date
total["reup"] = np.where(total['time'].str.contains("끌올"), 1, 0)
total['region_gu'] = total.region.str.extract(r"(\w+구)")
total
```

```python
sub = total[['num', 'price', 'the_date_ymd']].drop_duplicates()
# sub.sort_values(by='price', ascending=False).drop_duplicates(subset=['num', 'the_date_ymd'], keep='last')
sub
```
