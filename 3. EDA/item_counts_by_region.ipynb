```python
df = pd.read_excel("carrot_list_Seoul_선물세트_2021-09-25 140524.xlsx")
```

```python
df.head()
```

```python
df['gu'] = df.region.str.extract(r'([가-힣]+구)')
```

```python
df_amonth = df[df.time.str.contains("초|분|시간|일")]
```

```python
df_amonth.groupby('gu').title.count().sort_values(ascending=False).plot.bar(figsize=(12,5), fontsize = 12) 
# 구별 매물 수 
plt.ylabel('count', fontsize=15)
```

```python
df_amonth.groupby('gu').price.mean().sort_values(ascending=False).plot.bar(figsize=(12,5), fontsize = 12) 
# 구별 평균 가격 
plt.ylabel('AVG_price', fontsize=15)
```

```python
df_amonth.groupby('gu').price.median().sort_values(ascending=False).plot.bar(figsize=(12,5), fontsize = 12)
# 구별 중간 가격 
plt.ylabel('MEDIAN_price', fontsize=15)
```

```python
df_amonth.groupby('gu').view.mean().sort_values(ascending=False).plot.bar(figsize=(12,5), fontsize = 12)
# 구별 조회수 평균
plt.ylabel('AVG_view', fontsize=15)
```

```python
df_amonth.groupby('gu').interest.mean().sort_values(ascending=False).plot.bar(figsize=(12,5), fontsize = 12)
# 구별 관심 평균
plt.ylabel('AVG_interest', fontsize=15)
```
